```python
def cosine_similarity(query,document):
    #split them
    query_split = query.lower().split(" ")
    document_split = document.lower().split(" ")

    #put the result in dict[key:value] counter
    query_count = Counter(query_split)
    doc_count = Counter(document_split)
    mylist = []
    
    for tokens in query_count.keys() & doc_count.keys():
         mylist.append(query_count[tokens] * doc_count[tokens])
        
    dot_product = sum(mylist)

    #magnitude
    query_magnitude = math.sqrt(sum(query_count[token] ** 2 for token in query_count))
    document_magnitude = math.sqrt(sum(doc_count[token] ** 2 for token in doc_count))

    #calculate cosine_similarity
    similarity = dot_product/ (query_magnitude * document_magnitude)
    return similarity
    
    
```

```python
cosine_similarity(user_query,document)
```

```python
def return_response(query,corpus):
    similarities = []
    for doc in corpus:
        similarity = cosine_similarity(query,doc)
        similarities.append(similarity)
    return corpus_of_documents[similarities.index(max(similarities))]    
```

```python
import requests
import json
full_response = []
```

```python
full_response = []
prompt = """You are a bot that makes recommendation for different knowledge activities. You answer in very short sentence. Do not include
extra informations.
This is the recommendation documents: {relevant_documents}
The user input is: {user_input}
Compile a recommendation to the user based on the recommended activities and user input 
"""
url = 'http://localhost:11434/api/generate'
data =  {
  "model": "llama2",
  "prompt": prompt.format(user_input = user_input, relevant_documents= relevant_documents)
}

headers =  {'Content-Type': 'application/json'}
response = requests.post(url, data = json.dumps(data), headers=headers, stream=True)

try:
    for line in response.iter_lines():
        #filter out keep-alive new lines
        if line:
            decoded_line = json.loads(line.decode('utf-8'))
            full_response.append(decoded_line['response'])

finally:
    response.close()
    
print(''.join(full_response))                                  
                                      
```
